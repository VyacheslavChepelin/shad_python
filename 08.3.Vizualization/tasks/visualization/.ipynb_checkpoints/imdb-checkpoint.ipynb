### Домашнее задание по визуализации

В домашнем задании будем работать с данными фильмов на IMDB. 

Для начала считаем данные и проигнорируем все строки, которые не распарсились (для простоты).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('./imdb.csv', on_bad_lines='skip')
```

__Задача 1:__ С каким параметром больше всего коррелирует число побед? 
Ответ на задачу - название колонки в dataframe, с которой больше всего коррелирует `nrOfWins`.

```python
df = df.iloc[:, [10,5,6,7,8,11,12,13,14,15]]
corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', square=True)

plt.title('матрица корреляций')
plt.tight_layout()
plt.show()
```

__Задача 2:__ Посмотрите, как изменялась средняя оценка фильмов (рейтинг IMDB) с течением времени? В какое десятилетие была самая высокая средняя оценка. 

Ответ - tuple, где на первом месте тип тренда (слово 'ascending' или 'descending'), на втором - десятилие, например, если самые "хорошие" фильмы были в период 1950 - 1959, то 1950.

```python
df['decade'] = (df['year'] // 10) * 10
decade_ratings = df.groupby('decade')['imdbRating'].mean()
if decade_ratings.iloc[-1] > decade_ratings.iloc[0]:
    trend = 'ascending'
else:
    trend = 'descending'

best_decade = decade_ratings.idxmax()
result = (trend, int(best_decade))
print(result)
```

```python
df['decade'] = (df['year'] // 10) * 10
decade_ratings = df.groupby('decade')['imdbRating'].mean()

plt.figure(figsize=(12, 6))
decade_ratings.plot(kind='bar', color='skyblue')

plt.title('средний рейтинг IMDB по десятилетиям')
plt.xlabel('десятилетие')
plt.ylabel('средний рейтинг IMDB')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

__Задача 3:__ Исследумем распределение оценок фильмов по жанрам (если фильм принадлежит нескольким жанрам, то будем учитывать его в каждом из них). Для простоты будем смотреть на топ-10 самых популярных жанров (по числу фильмов с этим жанром). В ответе укажите 2 жанра: жанр с самой высокой средней оценкой (по медиане) и с самой большой дисперсией (по IQR). Если у нескольких жанров одинаковый показатель, то стоит указать первый по алфавиту. 

Ответ - tuple с двумя жанрами

```python
genre_columns = ['Action', 'Adult', 'Adventure', 'Animation', 'Biography', 'Comedy']
genre_columns.extend(['Crime', 'Documentary', 'Drama', 'Family', 'Fantasy', 'FilmNoir'])
genre_columns.extend(['GameShow', 'History', 'Horror', 'Music', 'Musical'])
genre_columns.extend(['Mystery', 'News', 'RealityTV', 'Romance', 'SciFi', 'Short'])
genre_columns.extend(['Sport', 'TalkShow', 'Thriller', 'War', 'Western'])


genre_counts = df[genre_columns].sum()
top_10_genres = genre_counts.nlargest(10).index.tolist()

median_ratings = []
iqr_ratings = []

for genre in top_10_genres:
    genre_movies = df[df[genre] == 1]['imdbRating'].dropna()

    median_ratings.append(genre_movies.median())
    Q1 = genre_movies.quantile(0.25)
    Q3 = genre_movies.quantile(0.75)
    iqr_ratings.append(Q3 - Q1)

genre_stats = pd.DataFrame({
    'genre': top_10_genres,
    'median_rating': median_ratings,
    'iqr': iqr_ratings
})

highest_median_genre = genre_stats.sort_values(['median_rating', 'genre'], ascending=[False, True]).iloc[0]['genre']
highest_iqr_genre = genre_stats.sort_values(['iqr', 'genre'], ascending=[False, True]).iloc[0]['genre']

result = (highest_median_genre, highest_iqr_genre)
print(result)
```

__Задача 4:__ Посмотрим на распределение фильмов по числу слов в названии. Какой график вы бы для этого построили и какова мода (самое частотое значение) этого распределения?

Ответ - число слов, которое встречается в большем количестве фильмов.

```python
df['word_count'] = df['title'].str.split().str.len()
plt.figure(figsize=(12, 6))
df['word_count'].hist(bins=30, alpha=0.7, edgecolor='black')
plt.title('распределение фильмов по количеству слов в названии')
plt.xlabel('количество слов в названии')
plt.ylabel('количество фильмов')
plt.grid(axis='y', alpha=0.3)
plt.show()

mode_word_count = df['word_count'].mode()[0]
print(mode_word_count)
```

__Вопрос 5:__ Посмотрим на то, как менялась со временем продолжительность фильмов? Постройте график и ответьте на вопрос, в каком десятилетии доля коротких фильмов (короче часа) была выше всего? Рассматриваем период >= 1920 года.

Ответ - десятилетие, например, если больше всего коротких фильмов было в период 1950 - 1959, то 1950.

```python
df_filtered = df[df['year'] >= 1920].copy()
df_filtered['decade'] = (df_filtered['year'] // 10) * 10
df_filtered['is_short'] = (df_filtered['duration'] < 3600).astype(int)

short_ratio = df_filtered.groupby('decade')['is_short'].mean()
max_decade = short_ratio.idxmax()

print(int(max_decade))
```

__Вопрос 6:__ И последний вопрос: фильмам какого жанра пользователи оставляют больше всего рецензий? Подумайте о  том, по какой метрике лучше всего смотреть. В ответе укажите жанр.

```python
genre_columns = ['Action', 'Adult', 'Adventure', 'Animation', 'Biography', 'Comedy']
genre_columns.extend(['Crime', 'Documentary', 'Drama', 'Family', 'Fantasy', 'FilmNoir'])
genre_columns.extend(['GameShow', 'History', 'Horror', 'Music', 'Musical'])
genre_columns.extend(['Mystery', 'News', 'RealityTV', 'Romance', 'SciFi', 'Short'])
genre_columns.extend(['Sport', 'TalkShow', 'Thriller', 'War', 'Western'])

reviews_by_genre = []
for genre in genre_columns:
    genre_movies = df[df[genre] == 1]
    avg_reviews = genre_movies['nrOfUserReviews'].mean()
    reviews_by_genre.append((genre, avg_reviews))

reviews_by_genre.sort(key=lambda x: x[1], reverse=True)
top_genre = reviews_by_genre[0][0]
print(top_genre)
```
